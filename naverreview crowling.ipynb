```python
import requests as req
from selenium import webdriver as wb #브라우저
from selenium.webdriver.common.keys import Keys #키보드
from selenium.webdriver.common.by import By #태그 소환툴
from bs4 import BeautifulSoup as bs
import time
from tqdm import tqdm
import pandas as pd
```

```python
driver = wb.Chrome() 
driver.get("https://rlp.kr/view.php?num=96200") #주소 &연산자 까지 지우면 쌩주소임
time.sleep(5)
body = driver.find_element(By.TAG_NAME, "body")
body.send_keys(Keys.END)
time.sleep(3)
```

```python
reviews2 = []
count = 0
```

```python
try:
    while True:
        for i in tqdm(range(1,8)):
            search = driver.find_element(By.XPATH,f"//*[@id=\"section_ajax_layer_2\"]/table/tbody/tr[{i}]/td/table[1]/tbody/tr/td[1]")
            soup = bs(search.text, "lxml")
            reviews2.append(soup.text)
        driver.find_element(By.CSS_SELECTOR,'#section_ajax_layer_2 > center > div.page_next_small.font_14.font_number').click()
        time.sleep(3)
        count += 1
        print(count)
except:
    print("에러")
```
